# IPL-score prediction data from 2008-2017abs

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import keras 
from sklearn import preprocessing 
# import tensorflow as tf 
```

```python
import tensorflow as tf 
```

```python
ipl= pd.read_csv("ipl_data.csv")
ipl.head(200)
```

```python
data = ipl.copy()
matches_per_venu = data[['mid','venue']].drop_duplicates()
matches_count = matches_per_venu['venue'].value_counts()

plt.figure(figsize=(12,6))
sns.barplot(x = matches_count.values,y = matches_count.index,palette="rainbow")
plt.title("Number of matches per venue")
plt.ylabel('Venue')
plt.show()
```

# in this next code block we will see individual score of a batsman

```python
runs_by_batsman = data.groupby('batsman')['runs'].max().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=runs_by_batsman.values, y=runs_by_batsman.index,palette="pastel")
plt.title('Top 10 Batsmen by Total Runs')
plt.xlabel('Total Runs')
plt.ylabel('Batsman')
plt.show()
```

```python
wicket_bowler = data.groupby('bowler')['wickets'].max().sort_values(ascending = False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(x = wicket_bowler.values,y= wicket_bowler.index,palette="rainbow")
plt.title("Top 10 wicket taker")
plt.xlabel("Total Wickets")
plt.ylabel("Bowler")
plt.show()
```

# Performing label encoding

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = ['bat_team', 'bowl_team', 'venue', "batsman", "bowler"]

data_encoded = data.copy()

label_encoders = {}

for col in cat_cols:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col])
    label_encoders[col] = le
```

```python
data_corr = data_encoded.drop(columns=["date","mid"],axis=1)
sns.heatmap(data_corr.corr(),annot=True)
plt.show()
```

# Splitting Data-sets

```python
feature_cols = ['bat_team', 'bowl_team', 'venue', 'runs', 'wickets', 'overs','striker','batsman','bowler']
X = data_encoded[feature_cols]
Y = data_encoded['total']

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

# Building a neural network

```python
model = keras.Sequential([
    keras.layers.Input( shape=(X_train_scaled.shape[1],)),
    keras.layers.Dense(512, activation='relu'),  
    keras.layers.Dense(216, activation='relu'),  
    keras.layers.Dense(1, activation='linear') 
])

huber_loss = tf.keras.losses.Huber(delta=1.0)  # You can adjust the 'delta' parameter as needed
model.compile(optimizer='adam', loss=huber_loss)
```

```python
model.fit(X_train_scaled,Y_train,epochs=10,batch_size=64,validation_data=(X_test_scaled,Y_test))
```

```python
model_losses = pd.DataFrame(model.history.history)
model_losses.plot()
```

# Evaluating the model

```python
predictions = model.predict(X_test_scaled)

from sklearn.metrics import mean_absolute_error,mean_squared_error
mean_absolute_error(Y_test,predictions)
```
